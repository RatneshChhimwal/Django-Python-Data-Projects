```python
# Distribution of Policy Expiry Dates - Analyze how policies are distributed across different expiration dates.
```

```python
expiration_distribution = df.groupby(df['Expiry']).size()
# the .size() method returns the number of elements in an object.
print(expiration_distribution)
```

```python
# Frequency of Different Locations** - Calculate the distribution of policies across urban and rural locations.
```

```python
location_frequency = df.groupby(df['Location']).size()
print(location_frequency)
```

```python
# **State-wise Policy Distribution** - Analyze the number of policies in each state (NY, WI, IL, NJ).
```

```python
state_policy_distribution = df.groupby(df['State']).size()
print(state_policy_distribution)
```

```python
# **Regional Policy Comparison** - Compare the number of policies between the East and Midwest regions.
```

```python
filtered_df = df[df['Region'].isin(['East','Midwest'])]
East_Midwest_Policy = filtered_df.groupby(filtered_df['Region']).size()
print(East_Midwest_Policy)
```
