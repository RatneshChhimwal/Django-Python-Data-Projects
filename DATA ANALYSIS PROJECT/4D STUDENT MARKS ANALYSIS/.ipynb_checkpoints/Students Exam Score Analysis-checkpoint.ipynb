```python
# As we can see, Our CSV file has an unnamed column at the index '0', So firsly, we drop the same
```

```python
df.drop(columns=[df.columns[0]], inplace=True)
```

```python
df.head(5)
```

```python
# As we can see, In out column 'WklyStudyHours', The values was supposed to be '5-10' and not '05-Oct', So we replace the values
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct','5-10')
```

```python
df.head(5)
```

```python
# TO KNOW THE COUNT OF NULL VALUES IN EACH COLUMN:
```

```python
df.isnull().sum()
```

```python
# Calculate total score for each student
df['TotalScore'] = df['MathScore'] + df['ReadingScore'] + df['WritingScore']

# Group by gender, calculate total score for each gender group, and sort by total score
TotalMarks_Gender = df.groupby('Gender', as_index=False)['TotalScore'].sum().sort_values(by='TotalScore', ascending=False)

# Optionally, drop the temporary TotalScore column
df.drop(columns='TotalScore', inplace=True)

# Display the result
print(TotalMarks_Gender)
```

```python
TotalMarks_Gender
```

```python
# Plot the bar plot
plt.figure(figsize=(10, 6))

TotalMarksXGender_plot = sns.barplot(data=TotalMarks_Gender, x='Gender', y='TotalScore', hue='Gender', palette='Set1', legend=False)

# hue='Gender' here specifies that the bars should be colored according to the 'Gender' column.

for bars in TotalMarksXGender_plot.containers:
    TotalMarksXGender_plot.bar_label(bars, fmt='%.0f')
    
plt.title('Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('TotalScore')
plt.show()
```

<h1 style=" font-size:18px; "> From the above chart we analysed that in total females have achieved more marks than male </h1>

# NOW, WE WANT TO ANALYSE THE IMPACT OF PARENT EDUCATION ON STUDENT MARKS

```python
Pe_marks = df.groupby('ParentEduc').agg({'MathScore':'mean', 
                                         'ReadingScore':'mean', 
                                         'WritingScore':'mean'})

# SYNTAX: 'DataFrame'.groupby('Column_Name').agg({Key_Column_Name: ['Column_Label', 'Aggregate_function(mean,sum,min,median,max,min,count)']})

# '.agg()' func is used to apply one or more aggregation operations to the columns of a DataFrame after grouping by one or more keys.
# '.agg()' takes a dictionary where the keys are the column names to be aggregated & the values are the aggregation functions to apply to those columns.

# IMPORTANT: To use the .agg() function effectively, you need to first group your DataFrame using groupby() based on one or more columns

Pe_marks
```

<h2 style= "font-size:18px;"> Now, We visualise the above analysis in a heat map</h2>

```python
plt.figure(figsize=(4,4))
sns.heatmap(Pe_marks, annot=True)      # SYNTAX: sns.heatmap(data, annot, linewidths, linecolor )
plt.show()
```

<h3 style=" font-size:18px; "> From the above chart we analysed that Parent-Education is directly proportional to the student average marks </h3>
