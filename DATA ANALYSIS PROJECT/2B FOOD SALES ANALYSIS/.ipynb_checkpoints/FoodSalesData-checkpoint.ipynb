```python
# 5. Analyze Sales Trends Over Time:** Plot a line chart showing how total sales vary over time (using the "Date" column). **************
```

```python
# FOR THIS REQUIREMENT, WE NEED TO SHOWCASE THE 'TotalRevenue' OVER THE PERIOD OF TIME 'DATE'.
# SO WE PUT 'DATE' & 'TOTAL_PRICE' on X&Y AXIS RESPECTIVELY
```

```python
Unique_dates = SampleDF['UnitPrice'].unique()
Unique_dates
```

```python
# Group by 'Date' and sum the 'TotalRevenue' for each date
sales_over_time = SampleDF.groupby('Date')['TotalRevenue'].sum()    # We group the sum of 'TotalRevenue' by 'Date' and assign it to var 'sales_over_time'

# Plotting the data

plt.figure(figsize=(10, 6))                                       # We use 'plt.figure(figsize=(width,height))' to creates a new figure for our plot
sales_over_time.plot(kind='line', marker='o', color='b')          # We use '.plot(kind='line')' function on 'sales_over_time' for line-chart
plt.title('Total Sales Over Time')                                # We use 'plt.title(TITLE)' to give a title to out plot
plt.xlabel('Date')                                                # We use'plt.xlabel('LABLE')' TO ASSIGN A LABEL TO X-AXIS
plt.ylabel('Total Sales')                                         # We use'plt.ylabel('LABLE')' TO ASSIGN A LABEL TO Y-AXIS
plt.grid(True, linewidth = 0.5)                                   # We use plt.grid(True/False) to add grids to out plot
plt.show()
```

```python
# 10.Perform Product Association Analysis:** Determine the average time it takes for each product to sell out before the next purchase
```

```python
SampleDF
```

```python
# Calculate the average restock time for each product
average_restock_time_per_product = SampleDF.groupby('Product')['Date'].diff().groupby(SampleDF['Product']).mean()

'''SampleDF.groupby('Product')['Date'].diff() : groups the DataFrame SampleDF by the 'Product' column 
and calculates the difference in dates ('Date' column) within each group. This results in a Series 
where each entry represents the time difference between consecutive dates for each product.

.groupby(SampleDF['Product']).mean() : further groups the Series obtained in the previous step 
by the 'Product' column and calculates the mean of the time differences for each product.
'''


# Display the average restock time for each product along with the product names
for product, avg_restock_time in average_restock_time_per_product.items():
    print(f"{product} - {avg_restock_time}")
```
